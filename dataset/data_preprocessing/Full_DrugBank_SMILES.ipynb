```python
wrong_smiles_repl_dict = {
    'DB00225': '[Gd+3].CNC(=O)CN(CCN(CCN(CC([O-])=O)CC(=O)NC)CC([O-])=O)CC([O-])=O',
    'DB00515': 'N.N.Cl[Pt]Cl',
    'DB00526': 'C1CC[C@H]([C@@H](C1)[NH-])[NH-].C(=O)(C(=O)O)O.[Pt+2]',
    'DB00958': 'C1CC(C1)(C(=O)O)C(=O)O.[NH2-].[NH2-].[Pt+2]'
}
```

```python
for index, row in df_smiles.iterrows():
    drugbank_id = row['DrugBank ID']
    if drugbank_id in list(missing_smiles_dict.keys()):
        df_smiles.loc[df_smiles['DrugBank ID'] == drugbank_id, 'SMILES'] = missing_smiles_dict[drugbank_id]
    elif drugbank_id in list(wrong_smiles_repl_dict.keys()):
        df_smiles.loc[df_smiles['DrugBank ID'] == drugbank_id, 'SMILES'] = wrong_smiles_repl_dict[drugbank_id]

df_smiles
```
