```python
from utils.utils import get_drug_index_from_text

def sliceindex(x):
    i = 0
    for c in x:
        if c.isalpha():
            i = i + 1
            return i
        i = i + 1
        
def upperfirst(x):
    i = sliceindex(x)
    return x[:i].upper() + x[i:]

drug_names["PubChem Compound ID"] = drug_names["CID000000085"].apply(lambda x: get_drug_index_from_text(x))
drug_names["name"] = drug_names["carnitine"].apply(lambda x: upperfirst(x))
drug_names.drop(columns=["CID000000085","carnitine"], inplace=True)
```

```python
final_df_1 = pd.merge(drug_links, common_drugs, how='left', on='PubChem Compound ID')
final_df_1 = final_df_1[['DrugBank ID', 'Name', 'PubChem Compound ID', 'name']]
final_df_1.rename(columns={'name':'STITCH Name'}, inplace=True)
final_df_1
```

```python
final_df_2[final_df_2["PubChem Compound ID_x"] == 119]
```

```python
final_df = final_df_2[(final_df_2["STITCH Name"].notna()) | (final_df_2["PubChem Compound ID_y"].notna()) |
                        (final_df_2["DrugBank ID"].isin(still_missing_drugbanks))]
```

```python
pd.set_option('display.max_rows', 20)
final_df.reset_index()
```

```python
final_df['PubChem Compound ID_y'].fillna(final_df['PubChem Compound ID_x'], inplace=True)
final_df['STITCH Name'].fillna(final_df['name'], inplace=True)
final_df.drop(columns=['name'], inplace=True)
```

```python
final_df.rename(columns={'Name':'DrugBank Name',
                        'PubChem Compound ID_x': 'DrugBank PubChem ID',
                        'PubChem Compound ID_y': 'STITCH ID'
                        }, inplace=True)
final_df
```

Add the missing drugs with DB13919 and DB04485, named 'Candesartan' and 'Thymidine', as well as the drug 'Levocarnitine' with DB00583 which is missing from the names csv

```python
Candesartan = pd.Series({'DrugBank ID': 'DB13919', 'DrugBank Name': 'Candesartan'})
Thymidine = pd.Series({'DrugBank ID': 'DB04485', 'DrugBank Name': 'Thymidine'})
Levocarnitine = pd.Series({'DrugBank ID': 'DB00583', 'DrugBank Name': 'Levocarnitine'})
```

```python
final_df = final_df.append(Candesartan, ignore_index=True)
final_df = final_df.append(Thymidine, ignore_index=True)
final_df = final_df.append(Levocarnitine, ignore_index=True)
final_df
```

```python
for drugbank, stitch in still_missing_drugs_dict.items():
    
    final_df.loc[final_df['DrugBank ID'] == drugbank, 'STITCH Name'] = stitch[1]
    final_df.loc[final_df['DrugBank ID'] == drugbank, 'STITCH ID'] = stitch[0]
```
